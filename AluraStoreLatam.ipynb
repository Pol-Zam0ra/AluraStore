### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
listaTiendas = [tienda,tienda2,tienda3,tienda4]
```

```python
for i, tienda in enumerate(listaTiendas, start=1):
    total = tienda['Precio'].sum()
    print(f'Ingreso total de la tienda {i}: {total}')
```

# 2. Ventas por categoría

```python
for i, tienda in enumerate(listaTiendas, start=1):
    categorias = tienda['Categoría del Producto'].value_counts()
    print(f'Cantidad de productos vendidos por categoría de la tienda {i}:')
    print(categorias)
```

# 3. Calificación promedio de la tienda


```python
for i, tienda in enumerate(listaTiendas, start=1):
    calificacion = tienda['Calificación'].mean()
    print(f'Calificación promedio de la tienda {i}: {calificacion}')
```

# 4. Productos más y menos vendidos




```python
import collections
for i, tienda in enumerate(listaTiendas, start=1):
    productoMasVendido = tienda['Producto'].value_counts().idxmax()
    ventasMasVendido = tienda['Producto'].value_counts().max()
    productoMenosVendido = tienda['Producto'].value_counts().idxmin()
    ventasMenosVendido = tienda['Producto'].value_counts().min()
    print(f'Producto más vendido de la tienda {i}:')
    print(f'Producto más vendido: {productoMasVendido} (Ventas: {ventasMasVendido})')
    print(f'Productos menos vendidos de la tienda {i}:')
    print(f'Producto menos vendido: {productoMenosVendido} (Ventas: {ventasMenosVendido})')
```

# 5. Envío promedio por tienda

```python
for i, tienda in enumerate(listaTiendas, start=1):
    promedioEnvio = tienda['Costo de envío'].mean()
    print(f'El costo promedio de la tienda {i}: {promedioEnvio}')
```

# 6. Graficos (Productos más vendidos,costos de envío y categoria de productos:)

```python
import matplotlib.pyplot as plt

for i, tienda in enumerate(listaTiendas, start=1):
    productosVendidos = tienda['Producto'].value_counts()

    plt.figure(figsize=(8, 6))
    productosVendidos.head(10).plot(kind='bar')  # Mostrar solo los 10 productos más vendidos
    plt.title(f'Productos más vendidos - Tienda {i}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad vendida')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
for i, tienda in enumerate(listaTiendas, start=1):
  tienda['Fecha de Compra'] = pd.to_datetime(tienda['Fecha de Compra'], dayfirst=True)

  costo_envio_mensual = tienda.groupby(tienda['Fecha de Compra'].dt.to_period('M'))['Costo de envío'].mean()
  costo_envio_mensual.index = costo_envio_mensual.index.to_timestamp()

  plt.figure(figsize=(10, 6))
  plt.plot(costo_envio_mensual.index, costo_envio_mensual.values, marker='o', linestyle='-')
  plt.title(f'Costo de Envío Promedio Mensual - Tienda {i}')
  plt.xlabel('Fecha')
  plt.ylabel('Costo de Envío Promedio')
  plt.grid(True)
  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()

```

```python
for i, tienda in enumerate(listaTiendas, start=1):
  categorias = tienda['Categoría del Producto'].value_counts()
  plt.figure(figsize=(8, 6))
  categorias.plot(kind='pie', autopct='%1.1f%%', startangle=90)
  plt.title(f'Distribución de Categorías - Tienda {i}')
  plt.ylabel('')  # Eliminar la etiqueta del eje y
  plt.show()
```
